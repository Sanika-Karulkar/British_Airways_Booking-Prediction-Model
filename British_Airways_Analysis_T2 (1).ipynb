```python
import numpy as np

def cap_outliers(df, column, round_to_int=False):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    if round_to_int:
        df[column] = df[column].round().astype(int)
    return df

# Apply capping to all numerical columns and round 'num_passengers' to the nearest integer
numerical_columns = ['num_passengers', 'purchase_lead', 'length_of_stay', 'flight_hour', 'flight_duration']
for col in numerical_columns:
    if col == 'num_passengers':
        df = cap_outliers(df, col, round_to_int=True)
    else:
        df = cap_outliers(df, col)
```

```python
import matplotlib.pyplot as plt
import numpy as np
def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")


plt.figure(dpi=100, figsize=(8, 5))
plot_mi_scores(mi_scores)
```

### XGBoost with SMOTE

```python
# Oversampling the Minority Class Using SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score, precision_score, f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import xgboost as xgb

# Oversampling the training data
smote = SMOTE()
x_train_resampled, y_train_resampled = smote.fit_resample(x_train, y_train)

# Feature scaling
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train_resampled)
x_test_scaled = scaler.transform(x_test)

# Model training with the resampled data
clf = xgb.XGBClassifier()
clf.fit(x_train_scaled, y_train_resampled)

# Predictions
predictions = clf.predict(x_test_scaled)

# Evaluation functions
def confusionmatrix(predictions):
    cm = confusion_matrix(y_test, predictions)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    return disp.plot()

def eval_classification(model):
    y_pred = model.predict(x_test_scaled)
    y_pred_train = model.predict(x_train_scaled)

    print("Accuracy (Test Set): %.2f" % accuracy_score(y_test, y_pred))
    print("Precision (Test Set): %.2f" % precision_score(y_test, y_pred))
    print("Recall (Test Set): %.2f" % recall_score(y_test, y_pred))
    print("F1-Score (Test Set): %.2f" % f1_score(y_test, y_pred))
    print("roc_auc (Test Set): %.2f" % roc_auc_score(y_test, y_pred))
    print("roc_auc (Train Set): %.2f" % roc_auc_score(y_train_resampled, y_pred_train))

    print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Evaluate the model
confusionmatrix(predictions)
eval_classification(clf)
```

Metrics- <br>
Precision: This metric indicates how many of the predicted bookings are actually complete. Higher precision reduces false positives. The XGBoost model with SMOTE has a higher precision (0.45) compared to the Random Forest (0.39).<br>
Recall: This metric shows how many actual bookings are correctly identified. Higher recall reduces false negatives. The Random Forest model with SMOTE has a higher recall (0.31) compared to the XGBoost model (0.18).<br>
F1-Score: This is the harmonic mean of precision and recall, providing a single metric that balances both. The Random Forest model has a higher F1-Score (0.35) compared to the XGBoost model (0.26).<br>

Findings- <br>
Random Forest with SMOTE has a better F1-Score and recall, making it more effective at identifying the minority class (booking complete) while maintaining a balanced performance.<br>

Conclusion-<br>
Considering the need to identify both booking completions (minority class) and non-completions (majority class) effectively, the Random Forest with SMOTE seems to be the better choice. It provides a better balance with a higher F1-Score and recall, ensuring that you capture more actual bookings while maintaining a reasonable level of precision.<br>
